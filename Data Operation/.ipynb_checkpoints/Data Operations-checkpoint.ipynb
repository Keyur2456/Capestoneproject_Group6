```python
# creating dataframe
mat = np.zeros((len(symptoms_data), len(symptoms)+1))

for i in range(len(symptoms_data)):
    mat[i,0] = diseases.index(symptoms_data.iloc[i,0])
    j = 1
    for j in range(1, 18):
        if symptoms_data.iloc[i,j] is not np.nan:
            sym_idx = symptoms.index(symptoms_data.iloc[i,j].replace(' ',''))
            mat[i,sym_idx+1] = 1    

symptoms_li = symptoms.copy()
symptoms_li.insert(0, 'Disease')      
di_sym_df = pd.DataFrame(mat, columns = symptoms_li)
```

```python
# construct adjaciency matrix
adj_mat = np.zeros((133,41))
for i in range(len(symptoms_data)):
    for j in range(1, 18):
        disease = symptoms_data.iloc[i,0]
        symptom = symptoms_data.iloc[i,j]
        if pd.notnull(symptom):
            symptom = symptom.replace(' ','')
            dis_index = diseases.index(disease)
            sym_index = symptoms.index(symptom)
            adj_mat[sym_index, dis_index] += 1
        else:
            pass
```

```python
plt.bar(np.arange(1,42),adj_mat.sum(axis=0))
plt.title('Probability distribution of diseases')
```
